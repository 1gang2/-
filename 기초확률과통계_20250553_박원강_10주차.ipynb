```python
import matplotlib.pyplot as plt

joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

f_X = {}
f_Y = {}

for x, y in joint_pmf_data.keys():
    f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]

for x, y in joint_pmf_data.keys():
    f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

plt.figure(figsize=(12, 5))

y_min = 0
y_max = 1

x_min = 0.5
x_max = 3.5

plt.subplot(1, 2, 1)
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo')
plt.vlines(list(f_X.keys()), y_min, list(f_X.values()), colors='blue')
plt.title('Marginal PMF of X')
plt.xlabel('X Value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max)+1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.subplot(1, 2, 2)
plt.plot(list(f_Y.keys()), list(f_Y.values()), 'ro')
plt.vlines(list(f_Y.keys()), y_min, list(f_Y.values()), colors='red')
plt.title('Marginal PMF of Y')
plt.xlabel('Y Value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max)+1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

data = np.array([korean_scores, math_scores])

covariance = np.cov(data, bias=True)[0, 1]

print("Covariance:", covariance)
```
